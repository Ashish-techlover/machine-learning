```python
import pandas as pd
import numpy as np
dt=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/lung-cancer/lung-cancer.data')
dt
```

```python
cols=['class','a1','a2','a3','a4','a5','a6','a7','a8','a9','a10','a11','a12','a13','a14','a15','a16','a17','a18','a19','a20','a21','a22','a23','a24','a25','a26','a27','a28','a29','a30','a31','a32','a33','a34','a35','a36','a37','a38','a39','a40','a41','a42','a43','a44','a45','a46','a47','a48','a49','a50','51','52','53','54','55','56']
dt.columns=[i for i in cols]
dt
```

```python
dt.a4=dt.a4.replace('?',-1)
dt.a38=dt.a38.replace('?',4)
dt,dt['a38']
```

```python
dt1=dt.drop('class',axis=1)
x=np.array(dt1)
y=np.array(dt['class'])
x,y
```

```python
#from sklearn.decomposition import PCA
#al=PCA(n_components=56)
#al.fit(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.85,test_size=0.15)
from sklearn.neighbors import KNeighborsClassifier
algo=KNeighborsClassifier(20)
algo.fit(x_train,y_train)
algo.score(x_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
k_range=range(1,20)
scores={}
scores_list=[]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2)
from sklearn.neighbors import KNeighborsClassifier
for k in k_range:
    algo=KNeighborsClassifier(n_neighbors=k)
    algo.fit(x_train,y_train)
    y_pred=algo.predict(x_test)
    scores[k]=metrics.accuracy_score(y_test,y_pred)
    scores_list.append(metrics.accuracy_score(y_test,y_pred))
scores,scores_list
```

```python
n=len(scores_list)
```

```python
s=0
for i in scores_list:
    s=s+i
print(s/n)
```
