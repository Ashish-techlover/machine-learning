```python
T=int(input())
t1=1
while t1<=T:
    tom_dir=[]
    jerry_dir=[]
    tom=[]
    jerry=[]
    range_of_inputs=int(input())
    N=int(input())
    for x in range(2):
        tom.append(int(input()))
    for x in range(2):
        jerry.append(int(input()))
    for l in range(1,range_of_inputs+1):
        tom_dir.append(input())
    for k in range(1,range_of_inputs+1):
        jerry_dir.append(input())
    if tom<=[N,N]:   
        for t in tom_dir:
            if t=='R' or t=='r':
                tom[1]=tom[1]+1
            elif t=='D' or t=='d':
                tom[0]=tom[0]+1
            elif t=='T' or t=='t':
                tom[0]=tom[0]-1
            elif t=='L' or t=='l':
                tom[1]=tom[1]-1
            else:
                continue
    if jerry<=[N,N]:
        for j in jerry_dir:
            if j=='R' or j=='r':
                jerry[1]=jerry[1]+1
            elif j=='D' or j=='d':
                jerry[0]=jerry[0]+1
            elif j=='T' or j=='t':
                jerry[0]=jerry[0]-1
            elif j=='L' or j=='l':
                jerry[1]=jerry[1]-1
            else:
                continue    
    if tom==jerry:
        print('DIED\n')
    else:
        print('ALIVE\n')
    t1=t1+1      
```
