```python
import pandas as pd
import numpy as np
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data')
```

```python
data.head()
```

```python
names=['buying','maint','doors','persons','lug_boot','safety','Class']
data.columns=[i for i in names]
data.columns
```

```python
data.buying=data.buying.replace('vhigh',1)
data.buying=data.buying.replace('high',2)
data.buying=data.buying.replace('med',3)
data.buying=data.buying.replace('low',4)
data.buying=data.buying.astype(dtype='i')
data.maint=data.maint.replace('vhigh',1)
data.maint=data.maint.replace('high',2)
data.maint=data.maint.replace('med',3)
data.maint=data.maint.replace('low',4)
data.maint=data.maint.astype(dtype='i')
```

```python
data.doors=data.doors.replace('5more',5)
data.doors=data.doors.astype(dtype='i')
```

```python
data.persons=data.persons.replace('more',6)
data.persons=data.persons.astype(dtype='i')
data.lug_boot=data.lug_boot.replace('small',1)
data.lug_boot=data.lug_boot.replace('med',2)
data.lug_boot=data.lug_boot.replace('big',3)
data.lug_boot=data.lug_boot.astype(dtype='i')
```

```python
data.safety=data.safety.replace('low',1)
data.safety=data.safety.replace('med',2)
data.safety=data.safety.replace('high',3)
data.safety=data.safety.astype(dtype='i')
```

```python
data.Class=data.Class.replace('unacc',1)
data.Class=data.Class.replace('acc',2)
data.Class=data.Class.replace('good',3)
data.Class=data.Class.replace('vgood',4)
data.Class=data.Class.astype(dtype='i')
```

```python
x=data.drop(['Class'],axis=1)
y=data['Class']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2)
from sklearn.neighbors import KNeighborsClassifier
algo=KNeighborsClassifier()
algo.fit(x_train,y_train)
print(algo.score(x_test,y_test))
```
