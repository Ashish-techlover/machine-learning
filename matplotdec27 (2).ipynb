```python
import matplotlib.pyplot as plt
```

```python
import numpy as n
```

```python
plt.style.use('ggplot')
a=n.linspace(-1,1,100)
plt.plot(a,n.arccos(a),':',color='b',label='cosinverse')
plt.plot(a,n.cos(a),'-.',color='g',label='cos')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("sin graph")
plt.legend()
plt.show()
```

```python
a=[1,2,3,4,5,6]
b=[3,5,4,7,6,8]
plt.scatter(a,b,marker='x',s=150,alpha=0.9,color='b',label='scatter')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("graph")
plt.legend()
plt.show()
```

```python
p=[1,1,1,1,2,2,2,2,2,2,3,4,5,6,7,8]
r=[1,2,3,4]
plt.hist(p,r,label='hist')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("graph")
plt.legend()
plt.show()
```

```python
x=['child','teen','adult','mature','old age']
y=[12,11,16,13,9]
x1=[0,4,2,3,1]
plt.bar(x1,y,tick_label=x)
#plt.xticks(x1,x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("graph")
#plt.legend()
plt.show()
```

```python
x=['child','teen','adult','mature','old age']
x1=[0,1,2,3,4]
y=[12,11,16,13,9]
y1=[3,7,6,5,2]
y2=[14,12,9,4,5]
plt.stackplot(x1,y,y1,y2,labels=x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("graph")
plt.legend()
plt.show()
```

```python
x=['child','teen','adult','mature','old age']
y=[12,11,16,13,9]
plt.pie(y,labels=x,autopct='%0.2f',explode=(0.1,0,0,0.2,0),shadow=True)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("graph")
#plt.legend()
plt.show()
```

```python
x=['child','teen','adult','mature','old age']
x1=[0,1,2,3,4]
y=[12,11,16,13,9]
p=plt.subplot(221)
p.scatter(x1,y)
p1=plt.subplot(222)
p1.bar(x1,y,tick_label=x)
p2=plt.subplot(223)
p2.pie(y,labels=x)
p3=plt.subplot(224)
p3.plot(x1,y)
plt.show()
```

```python
p=plt.subplot(111)
p.plot(x1,y)
p.annotate("best point",(1,11),xytext=(0.1,0.7),textcoords="axes fraction",arrowprops=dict(facecolor="black",color='blue'))
p.annotate("ideal point",(2,16),xytext=(0.7,0.1),textcoords="axes fraction",arrowprops=dict(facecolor="brown",color='blue'))
plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
f=plt.figure()
p=f.add_subplot(111,projection='3d')
p.plot(x1,y)
plt.show()
```

```python
f=plt.figure()
p=f.add_subplot(1,1,1,projection='3d')
p.bar3d(x1,y,n.zeros(5),n.ones(5),n.ones(5),x1)
plt.show()
```
