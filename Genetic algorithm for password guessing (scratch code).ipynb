# Creating generate_chromosome func. and generate_population func.

```python
import random
def generate_chromosome(target,geneset):
    length=len(target)
    genes=''
    while len(genes)<length:
        genes="".join(random.sample(geneset,length))#random.sample() returns list of samples
    return genes
def generate_population(target,geneset,size):
    population=[]
    for i in range(size):
        population.append((generate_chromosome(target,geneset)))
    return population
        
```

# Creating fitness function

```python
def fitness(population,target):
    s=0
    for i,j in zip(target,population):
        if i==j:
            s+=1
    return s
```

# Creating Crossover function

```python
def Crossover(population,fitness,target):#crossover function
    fit_list=[]
    new_genome=[]
    index=[]
    for i in range(len(population)):
        fit_list.append((fitness(population[i],target),population[i]))#elitism selection method
        fit_list.sort(reverse=True)#elitism selection method
    for j in range(len(fit_list)):
        new_genome.append(fit_list[j][1])#generating new genome in descending order of the fitness
    #print(new_genome)
    size=int((90*len(target))/100)#cutting point
    rate=random.randint(2,len(new_genome[:size]))#random crossover rate selection 
    parent=random.sample(new_genome[:size],rate)#selection of top 50% of fittest genome randomly
    for i in range(len(parent)):#searching for index values of the randomly selected geneset for later appending in new genome on same index value
        for j in range(len(new_genome)):
            if parent[i]==new_genome[j]:
                index.append(j)
    for i in range(len(parent)):
        if i==len(parent)-1:
            parent[i]=parent[i][:size]+parent[i-(len(parent)-1)][size:]
        else:
            parent[i]=parent[i][:size]+parent[i+1][size:]
    #print(parent)       
    for i,j in zip(index,parent):
        new_genome[i]=j
    return new_genome    
```

# Creating mutation function

```python
from math import ceil,floor
import random
def mutation(new_parents,geneset,rate):
    actual_index=[]
    new_child=[]
    total_gen=rate*len(new_parents)*len(new_parents[0])#total number of genes in population
    mr=floor(total_gen)#mutation rate for selecting number of mutations
    child=random.sample(new_parents,mr)
    for chrom in child:
        index=random.randrange(0,len(chrom))
        child_genes=list(chrom)
        mutation_gene,alternate_gene=random.sample(geneset,2)
        if child_genes[index]==mutation_gene:
            child_genes[index]=alternate_gene
            new_child.append(''.join(child_genes))
        else:
            child_genes[index]=mutation_gene
            new_child.append(''.join(child_genes))
    for i in range(len(child)):
        for j in range(len(new_parents)):
            if child[i]==new_parents[j]:
                actual_index.append(j)    
    for i,j in zip(actual_index,new_child):
        new_parents[i]=j
    return new_parents
```

# Creating GetOptimised function

```python
import datetime
def GetOptimised(geneset,population,target,fitness,crossover,mutation,mutation_rate):
    startTime = datetime.datetime.now()
    optimal_fitness=len(target)
    def show(population):
        bestfitness=[]
        timeDiff = datetime.datetime.now() - startTime
        for chromosome in population:
            bestfitness.append((fitness(chromosome,target),chromosome))
            bestfitness.sort(reverse=True)
        score,chromosome=bestfitness[0]#taking best fit value
        print("{0}\t{1}\t{2}".format(chromosome, score, str(timeDiff)))
        #BestFitness= max(bestfitness.append(score))
        return score,chromosome
    
    BestFitness,chromosome=show(population)
    if BestFitness>=optimal_fitness:
        return chromosome
    generation=0
    while True:
        generation+=1
        print('generation--> ',generation)
        parent=Crossover(population,fitness,target)
        child=mutation(parent,geneset,mutation_rate)
        childfitness, chromosome=show(child)
        if childfitness <= BestFitness:
            continue
        if childfitness>=optimal_fitness:
            break
        population=child
        BestFitness=childfitness
        
        
```

# Calling GetOptimised Function

```python
mutation_rate=0.045

GetOptimised(geneset,population,target,fitness,Crossover,mutation,mutation_rate)
```
