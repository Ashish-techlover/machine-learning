```python
#data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data')
#data.to_csv('e:\car.data')
data=pd.read_csv('F:\car.data')
data.dtypes
```

```python
data.head()
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data
```

```python
attr=['buying', 'maint' , 'doors','persons','lug_boot' ,'safety']+['class']
data.columns=[i for i in attr]
```

```python
data.head()
```

```python
data=data.replace('vhigh',1)
data=data.replace('high',2)
data=data.replace('med',3)
data=data.replace('low',4)
data=data.replace('5more',6)
data=data.replace('more',5)
data=data.replace('small',1)
data=data.replace('med',3)
data=data.replace('big',2)
data=data.replace('unacc',1)
data=data.replace('acc',2)
data=data.replace('good',3)
data=data.replace('vgood',4)
```

```python
car = data.values
```

```python
x,y=car[:,:6],car[:,6]
```

```python
#car.astype(int)
x,y = x.astype(int), y.astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2)
from sklearn.neighbors import KNeighborsClassifier
algo=KNeighborsClassifier()
algo.fit(x_train,y_train)
#print(algo.predict([x]))#requires dimension less tha equals to 2
print(algo.score(x_test,y_test))
```
